```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 10, 500)
true_signal = np.sin(time)
```

```python
noise = np.random.normal(0, 0.3, size=time.shape)
sensor_measurement = true_signal + noise
```

```python
plt.figure(figsize=(8,4))
plt.plot(time, true_signal, label="True Signal")
plt.plot(time, sensor_measurement, label="Sensor Measurement", alpha=0.7)
plt.legend()
plt.title("Simulated Sensor Data with Noise")
plt.show()
```

```python
extracted_noise = sensor_measurement - true_signal

plt.figure(figsize=(8,4))
plt.plot(time, extracted_noise, color="purple")
plt.title("Extracted Noise from Sensor Measurement")
plt.xlabel("Time")
plt.ylabel("Noise")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(extracted_noise, bins=40)
plt.title("Noise Distribution")
plt.xlabel("Noise Value")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Noise Mean:", np.mean(extracted_noise))
print("Noise Standard Deviation:", np.std(extracted_noise))
```
